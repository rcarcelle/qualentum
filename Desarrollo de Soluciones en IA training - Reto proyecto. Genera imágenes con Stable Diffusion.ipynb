```python
from diffusers import StableDiffusionPipeline, EulerAncestralDiscreteScheduler
import torch
import os

# Usa GPU si está disponible
device = "cuda" if torch.cuda.is_available() else "cpu"

# Carga el scheduler por defecto
scheduler = EulerAncestralDiscreteScheduler.from_pretrained(
    "stabilityai/stable-diffusion-2-1-base", subfolder="scheduler"
)

# Carga el pipeline de Stable Diffusion
pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1-base",
    scheduler=scheduler,
    torch_dtype=torch.float16
).to(device)

# Carpeta para guardar imágenes generadas
output_dir = "/content/drive/MyDrive/Colab Notebooks/generated_images"
os.makedirs(output_dir, exist_ok=True)
```

```python
prompts = [
    "A sleek red sports car parked on a mountain road at sunset, photorealistic",
    "A futuristic electric car with glowing blue accents driving through a neon-lit city at night",
    "A vintage classic car in pristine condition, displayed at an outdoor car show",
    "A rugged off-road SUV splashing through a muddy trail in a forest, dynamic action shot",
    "A luxury black sedan with chrome details parked in front of a modern skyscraper, studio lighting"
]

for i, prompt in enumerate(prompts):
    image = pipe(prompt, guidance_scale=7.5, num_inference_steps=50).images[0]
    image.save(f"{output_dir}/generated_image_{i+1}.png")
```

```python
from diffusers import StableDiffusionImg2ImgPipeline, LMSDiscreteScheduler
from PIL import Image

# Carga una imagen base de custom_dataset (asegúrate que la ruta y nombre sean correctos)
input_image_path = "/content/drive/MyDrive/Colab Notebooks/custom_dataset/16805217155714.jpg"  # Cambia por tu imagen
init_image = Image.open(input_image_path).convert("RGB")
init_image = init_image.resize((512, 512))  # redimensionar a 512x512

# Carga otro scheduler para image-to-image
img2img_scheduler = LMSDiscreteScheduler.from_pretrained(
    "stabilityai/stable-diffusion-2-1-base", subfolder="scheduler"
)

img2img_pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1-base",
    scheduler=img2img_scheduler,
    torch_dtype=torch.float16
).to(device)

variation_prompts = [
    "A futuristic version of the car with neon lights and holographic details",
    "A classic car transformed into a cyberpunk style with glowing accents",
    "An off-road vehicle upgraded for a post-apocalyptic environment with heavy armor",
    "A luxury car with a matte black finish and red racing stripes under dramatic lighting",
    "A sports car redesigned as a concept vehicle with aerodynamic shapes and bright colors"
]

strength_values = [0.5, 0.8]  # cuánto cambiará la imagen base

# Carpeta para guardar variaciones
variations_dir = os.path.join(output_dir, "variations")
os.makedirs(variations_dir, exist_ok=True)

for i, (prompt, strength) in enumerate(zip(variation_prompts, strength_values)):
    generated_img = img2img_pipe(
        prompt=prompt,
        image=init_image,
        strength=strength,
        guidance_scale=7.5,
        num_inference_steps=50
    ).images[0]
    generated_img.save(f"{variations_dir}/variation_{i+1}_strength_{strength}.png")
```
